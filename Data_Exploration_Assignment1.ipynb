```python
# 3. Data Manipulation with Pandas
# Create sample dataset using a dictionary
# Convert into a Pandas dataFrame and display output

import pandas as pd

# Dataset
data = {
    "Name": ["Kodi", "Keilah", "Kacee"],
    "Age": [8, 7, 5],
    "City": ["Toronto", "North York", "Scarborough"]
}

df = pd.DataFrame(data)
df

```

# Add a Markdown cell to describe the data

# Dataset Description
This dataset contains fictional information about three individuals, including their names, ages, and location.


```python
# 4. Data Visualization with Matplotlib/Seaborn
# Bar Chart

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart visualization
plt.figure(figsize=(6,4))
sns.barplot(x="Name", y="Age", data=df)
plt.title("Age Distribution by Name")
plt.xlabel("Name")
plt.ylabel("Age")
plt.tight_layout()
plt.show()
```

# Markdown Cell: Explanation of the Visualization
